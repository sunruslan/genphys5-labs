## Измерение коэффициента ослабления потока $\gamma$-лучей в веществе и определение их энергии
*С помощью сцинтилляционного счетчика измеряются линейные коэффициенты ослабления потока $\gamma$-лучей в свинце, железе и алюминии; по их величине определяется энергия $\gamma$-квантов.*
### Теоретическая справка
Проходя через вещество, поток $\gamma$-квантов постепенно ослабляется. 
$$I = I_0e^{-\mu l} = I_0e^{-\mu'm_1}$$
$\mu'$ в отличие от $\mu$ не зависит от плотности среды, т.е. при увеличении плотности в два раза $\mu$ увеличится в два раза. 
#### Фотоэлектрическое поглощение
Свободный электрон не может поглотить или испустить фотон.
Вероятность фотоэффекта пропорциональна $Z^5$. 
Возникает при не очень больших энергиях.

#### Комптоновское рассеяние
Происходит на слабосвязанных электронах.

#### Образование пар
При энергиях $\gamma$-лучей, превышающих $2mc^2=1.02\text{ МэВ}$. Вероятность возникновения электрон-позитронных пар пропорциональна $Z^2$.

#### Полный коэффициент ослабления потока $\gamma$-лучей
Равна сумме коэффициентов для всех трех рассмотренных процессов.

$$\mu = \frac{1}{l}\ln{\frac{N_0}{N}}$$

### Вывод
Измерили коэффициенты ослабления потока $\gamma$-лучей в свинце, железе и алюминии; по их величине определяется энергия $\gamma$-квантов.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = None
I0 = 25743.8
b = [1.95, 1, 0.47]
data = [pd.read_csv('datasetAl', sep=' ', header=None), pd.read_csv('datasetFe', sep=' ', header=None), pd.read_csv('datasetPb', sep=' ', header=None)]
for i in range(3):
    data[i].columns = [r'$\text{N, количество частиц}$', r'$\text{t, с}$'];
    data[i][r'$\epsilon, \%$'] = np.sqrt(data[i].iloc[:, 0]) / data[i].iloc[:, 0] * 100
    data[i][r'$\text{b, см}$'] = b[i]*np.arange(1, data[i].shape[0]+1)
    data[i][r'$\text{I, шт/сек}$'] = data[i].iloc[:, 0] / data[i].iloc[:, 1]
    data[i][r'$\ln{\frac{I_0}{I}}$'] = np.log((data[i].iloc[:, 4]-3046/250)/ (I0-3046/250))*(-1)
display(data[0])
display(data[1])
display(data[2])
```

```python
plt.figure(figsize=(10, 10))
plt.legend()
labels = [u'Алюминий', u'Железо', u'Свинец']
for i in (0, 1, 2):
    x = [0]+list(data[i].iloc[:, 3])
    y = [0]+list(data[i].iloc[:, -1])
    A = np.vstack([x, np.ones(len(x))]).T
    m,c = np.linalg.lstsq(A, y, rcond=None)[0]
    f = lambda x: m*x+c
    print(m)
    xx = np.linspace(0, np.max(data[i].iloc[:, 3]))
    plt.plot(xx, f(xx), label=labels[i])
    plt.errorbar([0]+list(data[i].iloc[:, 3]), [0]+list(data[i].iloc[:, -1]), 
                 xerr=[0]+list(0.05*np.ones(data[i].shape[0])),
                 yerr=[0]+list(np.sqrt((1./np.sqrt(257438))**2+(data[i].iloc[:, 3]/100)**2)), 
                 fmt='none')
plt.legend()
plt.grid(b=True, which='major', color='k', linestyle='-')
plt.grid(b=True, which='minor', color='r', linestyle='--', alpha=0.75)
plt.xlabel(u'$b, см$', fontsize=24)
plt.ylabel(r'$\ln{\frac{I_0}{I}}$', fontsize=24)
plt.minorticks_on()
plt.savefig('gra')
plt.show()
```
